# 🎨 LayoutGen: Stable Diffusion on Google Colab

This notebook allows you to generate images from text prompts (derived from layouts) using Stable Diffusion v1.5 with GPU acceleration.

```python
# @title 1. Install Dependencies
!pip install diffusers transformers accelerate ftfy
```

```python
# @title 2. Load Stable Diffusion Model
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import os

model_id = "runwayml/stable-diffusion-v1-5"
device = "cuda"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)

print("Model loaded successfully!")
```

```python
# @title 3. Generate Image
prompt = "A serene living room featuring a blue sofa at the top left, a wooden table at the bottom center." # @param {type:"string"}

if not os.path.exists("outputs"):
    os.makedirs("outputs")

print(f"Generating image for: {prompt}")
image = pipe(prompt).images[0]

image.save("outputs/generated_colab_image.png")
display(image)
```

### How to use with LayoutGen Streamlit app:
1. Copy the **Generated Prompt** from the Streamlit UI.
2. Paste it into the `prompt` field above.
3. Run the cell to generate the image using Colab's GPU.
